```python
from pyspark.sql import SparkSession
import socket

print("=" * 60)
print("TESTE DE CONEXÃO JUPYTER → SPARK")
print("=" * 60)

# 1. Teste de rede
print("\n1️⃣ TESTE DE REDE:")
print(f"   Hostname do Jupyter: {socket.gethostname()}")
print(f"   IP do Jupyter: {socket.gethostbyname(socket.gethostname())}")

try:
    ip_spark = socket.gethostbyname("spark-master")
    print(f"   ✓ DNS resolvido: spark-master → {ip_spark}")
except Exception as e:
    print(f"   ✗ Erro ao resolver DNS: {e}")

# Testa porta 7077
try:
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(5)
    result = sock.connect_ex(("spark-master", 7077))
    if result == 0:
        print(f"   ✓ Porta 7077 acessível!")
    else:
        print(f"   ✗ Porta 7077 NÃO acessível")
    sock.close()
except Exception as e:
    print(f"   ✗ Erro ao testar porta: {e}")

# 2. Cria SparkSession
print("\n2️⃣ CRIANDO SPARK SESSION:")
try:
    spark = (
        SparkSession.builder
        .appName("Teste de Conexão")
        .master("spark://spark-master:7077")
        .config("spark.driver.host", "spark_notebook")
        .config("spark.driver.bindAddress", "0.0.0.0")
        .getOrCreate()
    )
    print(f"   ✓ SparkSession criada!")
    print(f"   ✓ Versão: {spark.version}")
    print(f"   ✓ Master: {spark.sparkContext.master}")
    print(f"   ✓ App ID: {spark.sparkContext.applicationId}")
except Exception as e:
    print(f"   ✗ Erro ao criar session: {e}")
    exit()

# 3. Teste simples
print("\n3️⃣ TESTE DE OPERAÇÃO:")
try:
    data = [("João", 25), ("Maria", 30), ("Pedro", 35)]
    df = spark.createDataFrame(data, ["nome", "idade"])
    count = df.count()
    print(f"   ✓ DataFrame criado com sucesso!")
    print(f"   ✓ Registros: {count}")
    df.show()
except Exception as e:
    print(f"   ✗ Erro na operação: {e}")

print("\n" + "=" * 60)
print("✅ CONEXÃO OK! Pode seguir com seus scripts.")
print("=" * 60)
```
