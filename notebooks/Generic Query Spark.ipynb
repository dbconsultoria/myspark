```python
from pyspark.sql import SparkSession

def main():
    # Initialize Spark
    spark = (
        SparkSession.builder
        .appName("custom_query")
        .master("local[*]")
        .config("spark.hadoop.fs.defaultFS", "file:///")
        .getOrCreate()
    )
    
    # Load tables
    tables = {
        "dim_customers": "file:/data/gold/dim_customers",
        "dim_products": "file:/data/gold/dim_products",
        "dim_categories": "file:/data/gold/dim_categories",
        "fact_orders": "file:/data/gold/fact_orders"
    }
    
    for name, path in tables.items():
        try:
            spark.read.parquet(path).createOrReplaceTempView(name)
            print(f"✓ {name} loaded")
        except:
            print(f"✗ {name} failed")
    
    # ========================================
    # YOUR CUSTOM QUERY HERE - EDIT FREELY
    # ========================================
    
    query = """
        SELECT
            *
        FROM fact_orders f
        order by 1,2,3,4
    """
    
    # ========================================
    # Execute and show results
    # ========================================
    
    print("\n" + "="*80)
    print("QUERY RESULTS")
    print("="*80)
    
    result = spark.sql(query)
    result.show(100, truncate=False)
    
    print(f"\nTotal rows: {result.count()}")
    
    spark.stop()

if __name__ == "__main__":
    main()
```
