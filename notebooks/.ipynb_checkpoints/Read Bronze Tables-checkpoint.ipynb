```python
from pyspark.sql import SparkSession
import os

# Cria a sessão Spark
spark = (
    SparkSession.builder
    .appName("Listar Tabelas Bronze")
    .master("spark://spark-master:7077")
    .config("spark.driver.host", "spark_notebook")
    .config("spark.driver.bindAddress", "0.0.0.0")
    .getOrCreate()
)

print("=" * 60)
print("TABELAS NA CAMADA BRONZE")
print("=" * 60)

# Caminho base do bronze
bronze_path = "/data/bronze"

# Lista os bancos de dados (subpastas em bronze)
databases = []
for db in os.listdir(bronze_path):
    db_path = os.path.join(bronze_path, db)
    if os.path.isdir(db_path):
        databases.append(db)

print(f"\n✓ Encontrados {len(databases)} banco(s) de dados:\n")

# Para cada banco, lista as tabelas
total_tables = 0
for db in sorted(databases):
    db_path = os.path.join(bronze_path, db)
    tables = []
    
    for table in os.listdir(db_path):
        table_path = os.path.join(db_path, table)
        if os.path.isdir(table_path):
            tables.append(table)
    
    print(f"📁 {db}:")
    if tables:
        for table in sorted(tables):
            # Verifica se tem arquivos parquet
            table_path = os.path.join(db_path, table)
            files = [f for f in os.listdir(table_path) if f.endswith('.parquet')]
            file_count = len(files)
            print(f"   └─ {table} ({file_count} arquivo(s) parquet)")
            total_tables += 1
    else:
        print(f"   └─ (vazio)")
    print()

print("=" * 60)
print(f"TOTAL: {total_tables} tabela(s) encontrada(s)")
print("=" * 60)
```
